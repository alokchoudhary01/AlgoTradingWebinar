```python
# Import required library
import yfinance as yfin
import talib as ta
import matplotlib.pyplot as plt

# Getting data
nifty = yfin.download('^NSEI', '2008-1-1','2020-04-01')
```

```python
# SMA & EMA
nifty['Simple MA'] = ta.SMA(nifty['Close'],20)
nifty['EMA'] = ta.EMA(nifty['Close'], timeperiod = 20)
# Display Chart
nifty[['Close','Simple MA','EMA']].plot(figsize=(25,20))
plt.show()
```
